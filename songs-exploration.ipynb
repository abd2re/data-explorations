# Exploration of most popular songs from 1950s-2010s

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## The different decades

```python
decades = '1950 1960 1970 1980 1990 2000 2010'.split(" ")
```

```python
datasets = [[] for i in range(len(decades))]
```

# Pre-processing

```python
for i in range(len(decades)):
  datasets[i] = pd.read_csv(f"../input/spotify-past-decades-songs-50s10s/{decades[i]}.csv")
  datasets[i]['decade'] = decades[i]
```

```python
datasets[0].head()
```

```python
df = pd.concat(datasets)
```

```python
df
```

```python
means =  pd.concat([df[df['decade'] == i].describe().loc[['mean']].iloc[:,2:].transpose() for i in decades],axis=1)
std = pd.concat([df[df['decade'] == i].describe().loc[['std']].iloc[:,2:].transpose() for i in decades],axis=1)
```

```python
means.columns = decades
std.columns = decades
means, std = means.transpose(), std.transpose()
```

# Simple Data exploration

```python
def exploreData(datatype):
    fig,ax = plt.subplots()
    ax.plot(means[datatype], color="red", marker="o")
    ax.set_xlabel("year", fontsize = 14)
    ax.set_ylabel("values", color="red", fontsize=14)
    ax2=ax.twinx()
    ax2.plot(std[datatype],color="blue",marker="o")
    ax2.set_ylabel("volatility",color="blue",fontsize=14)
    ax.set_title(datatype)
    plt.show()
```

```python
exploreData('dB')
exploreData('bpm')
exploreData('dnce')
exploreData('val')
exploreData('nrgy')
exploreData('dur')
exploreData('spch')
exploreData('pop')
```
